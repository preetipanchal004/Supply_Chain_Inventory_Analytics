```python
region_revenue = df.groupby("Region")["Revenue"].sum().sort_values(ascending=False)
print(region_revenue)
```

```python
warehouse_revenue = df.groupby("Warehouse_ID")["Revenue"].sum().sort_values(ascending=False)
print(warehouse_revenue)
```

```python
        Warehouse2 was the best-performing warehouse.
```

```python
top_products = df.groupby("SKU_ID")["Units_Sold"].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
         The top-selling SKUs contribute significantly to overall sales.
```

```python
promotion = df.groupby("Promotion_Flag")["Revenue"].sum()
print(promotion)
```

```python
         Promotion orders contributed less revenue than non-promotion orders.
```

```python
inventory = df.groupby("Warehouse_ID")["Inventory_Level"].mean()
print(inventory)
```

```python
        warehouse 2 has the highest average inventory level.
```

```python
import matplotlib.pyplot as plt
```

```python
region_revenue = df.groupby("Region")["Revenue"].sum()
ax=region_revenue.plot(kind="bar")
for container in ax.containers:
    ax.bar_label(container,fmt='%.2f')
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.show()
```

```python
warwhouse_revenue = df.groupby("Warehouse_ID")["Revenue"].sum()
bx=warehouse_revenue.plot(kind="bar")
for container in bx.containers:
    bx.bar_label(container,fmt='%.2f')
plt.title("Revenue by Warehouse")
plt.xlabel("Warehouse")
plt.ylabel("Revenue")
plt.show()
```

```python
promotion = df.groupby("Promotion_Flag")["Revenue"].sum()
#promotion = df["Promotion_Flag"].value_counts()
promotion.plot(kind="pie",autopct="%1.1f%%")
plt.title("Promotion Distribution")
plt.show()
```

```python
plt.hist(df["Inventory_Level"],bins=20)
plt.title("Inventory Distribution")
plt.xlabel("Inventory Level")
plt.ylabel("Frequency")
plt.show()
```

```python
top_products = df.groupby("SKU_ID")["Units_Sold"].sum().sort_values(ascending=False).head(10)
top_products.plot(kind="bar")
plt.title("Top 10 Products")
plt.xlabel("SKU")
plt.ylabel("Units Sold")
plt.show()
```

```python
top_revenue = df.groupby("SKU_ID")["Revenue"].sum().sort_values(ascending=False).head(10)
print(top_revenue)
```

```python
top_supplier = df.groupby("Supplier_ID")["Profit"].sum().sort_values(ascending=False).head(10)
print(top_supplier)
```

```python
lead_time = df.groupby("Supplier_ID")["Supplier_Lead_Time_Days"].mean().sort_values(ascending=False)
print(lead_time)
```

```python
stockout = df.groupby("Stockout_Flag")["Revenue"].sum()
print(stockout)
```
